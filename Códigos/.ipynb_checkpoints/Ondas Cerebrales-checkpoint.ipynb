```python
x = EEG
dt=tiempo[1]-tiempo[0]
N = x.shape[0]
T = N*dt
```

```python
print(np.mean(x))
print(x.mean())
print(x.var())
print(x.std())
```

```python
lags= np.arange(-len(x)+1,len(x))
print(x.shape)
x = x.flatten()
print(x.shape)
ac = (1/N)*np.correlate(x-x.mean(),x-x.mean(),mode='full')
indices = abs(lags)<=100
```

```python
plt.figure(figsize=(12,3))
plt.plot(lags[indices]*dt,ac[indices])
plt.xlabel('Lags[s]')
plt.ylabel('Autocovarianza')
plt.title('EEG')
plt.show()
```

```python
from numpy.fft import fft,rfft

xf = fft(x-x.mean())
#Espectro de potencia
Sxx=(2*dt**2/T)*(xf*np.conj(xf))
#Ignorando las frecuencias negativas
Sxx=Sxx[:int(len(x)/2)]
#Resolución de la frecuencia
df=1/T
#Frecuencia de Nyquist
fNQ=1/dt/2
#Eje de frcuencia
eje_f=np.arange(0,fNQ,df)

plt.figure(figsize=(12,3))
plt.plot(eje_f,np.real(Sxx))
plt.xlabel('Frecuencia[Hz]')
#plt.ylabel('Potencia [$\muV^2$/Hz]')
plt.xlim([0,100])
plt.show()
```

Ahora escribimos en términos de decibeles
- En 60Hz, se sigue viendo el pico, ahora están un poco escalados los cambios y se pueden observar un par de picos que pasan los valores medios de -50 db.

```python
plt.figure(figsize=(12,3))

eje_x = eje_f
eje_y = 10*np.log10(Sxx/max(Sxx))

plt.plot(eje_x,eje_y)
plt.xlabel('Frecuencia[Hz]')
plt.ylabel('Potencia [dB]')

plt.xlim([0,100])
plt.ylim([-60,0])
plt.show()
```

Modificamos el eje x, quiero que la gráfica sea logaritmica en el eje x.
Espectro: Tomo una porción de la señal y en toda esa señal intento extraer cuáles son las frecuencias que dominan en la señal

Se usa la transformada rápida de Fourier, busca si la señal coincide con senos y cocenos a una cierta señal específica, cuando coinciden me regresa un valor alto. Entonces sé que mi señal está en esa frecuencia.

```python
plt.figure(figsize=(12,3))

eje_x = eje_f
eje_y = 10*np.log10(Sxx/max(Sxx))

plt.semilogx(eje_x,eje_y)
plt.xlabel('Frecuencia[Hz]')
plt.ylabel('Potencia [dB]')

plt.xlim([0,100])
plt.ylim([-60,0])
plt.show()
```

```python
# Espectograma

from scipy.signal import spectrogram

Fs = 1/dt #Frecuencia de muestreo
intervalo = int(Fs)
sobrelape = int(Fs*0.95)

f,t,Sxx = spectrogram(x,fs=Fs,nperseg=intervalo,noverlap=sobrelape)

plt.figure(figsize=(12,3))

eje_x = t
eje_y = f

plt.pcolormesh(eje_x,eje_y,10*np.log10(Sxx),cmap="jet")

plt.xlabel('Tiempo[s]')
plt.ylabel('Frecuencia [Hz]')
plt.ylim([0,100])
plt.show()
```
